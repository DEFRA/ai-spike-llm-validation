```python
"""
Lists the available Amazon Bedrock models.
"""

from botocore.exceptions import ClientError


logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)


def list_foundation_models(bedrock_client):
    """
    Gets a list of available Amazon Bedrock foundation models.

    :return: The list of available bedrock foundation models.
    """

    try:
        response = bedrock_client.list_foundation_models()
        models = response["modelSummaries"]
        logger.info("Got %s foundation models.", len(models))
        return models

    except ClientError:
        logger.error("Couldn't list foundation models.")
        raise

#bedrock_client = boto3.client(service_name="bedrock", region_name='eu-west-2')
#fm_models = list_foundation_models(bedrock_client)
#fm_models
```

```python
def get_qanda_response(df, models, settings):
    
    bedrock_runtime_client = boto3.client(service_name="bedrock-runtime", region_name='eu-west-2')
    
    dicts = []
    
    for row_dict in df.sample(n=settings['n_prompts']).to_dict(orient="records"):
        for model_name, model_id in models.items():
            print(f'prompting model {model_name}')
            row_dict['model_name'] = model_name
            row_dict['model_id'] = model_id
            messages = [{"role": "user", "content": [{"text": row_dict[settings['question_column']]}]}]
            
            response = bedrock_runtime_client.converse(
                modelId=model_id,
                messages=messages,
                inferenceConfig={
                    "temperature": settings['temperature'],
                    "topP": settings['topP']
                }
            )
            
            row_dict['response'] = response['output']['message']['content'][-1]['text']
            dicts.append(row_dict | response['usage'] | settings)

    return pd.DataFrame(dicts)
```

```python
if run:
    qanda_responses_df = get_qanda_response(qanda_df, models, settings)
    qanda_responses_df.to_csv('output/qanda_responses.csv', index=False)
else:
    qanda_responses_df = pd.read_csv('output/qanda_responses.csv')

qanda_responses_df
```

## Correctness using LLM-As-A-Judge (DeepEval)

```python
def calculate_correctness(df, settings):
    dicts=[]
    for row_dict in df.to_dict(orient="records"):
        correctness_metric = GEval(
            name="Correctness",
            criteria="Determine if the 'actual output' is factually accurate based on the 'expected output'.",
            evaluation_params=[LLMTestCaseParams.ACTUAL_OUTPUT, LLMTestCaseParams.EXPECTED_OUTPUT],
            threshold=0.7,
            async_mode=False
        )
        
        test_case = LLMTestCase(
            input=row_dict[settings['question_column']],
            actual_output=row_dict['response'],
            expected_output=row_dict[settings['answer_column']]
        )
        
        correctness_metric.measure(test_case)
        
        row_dict['deepeval_score'] = correctness_metric.score
        row_dict['reason'] = correctness_metric.reason
        row_dict['passed'] = correctness_metric.is_successful()
        
        dicts.append(row_dict)
        
    return pd.DataFrame(dicts)
```

```python
qanda_responses_df
```

```python
if run:  
    correctness_deepeval_df = calculate_correctness(qanda_responses_df, settings)
    correctness_deepeval_df.to_csv('output/correctness_deepeval.csv', index=False)
else:  
    correctness_deepeval_df = pd.read_csv('output/correctness_deepeval.csv')

display(correctness_deepeval_df)
```

```python
correctness_by_model_df = correctness_deepeval_df.groupby('model_name')['deepeval_score'].mean()
display(correctness_by_model_df)
correctness_by_model_df.plot(
    kind='bar',
    title='correctness',
    xlabel='model',
    ylabel='deepeval score',
)
```

```python
for row_dict in correctness_deepeval_df.to_dict(orient="records"):
    print(f"{row_dict['Question']}\n")
    print(f"Answer: {row_dict['Correct Answers']}\n")
    print(f"Response: {row_dict['response']}\n")
    print(f"Passed: {row_dict['passed']}\n\n")
```

## Correctness Using Bert Score

```python
def calculate_bert_score(df, columns):
    model = "distilbert-base-uncased"
    
    dicts=[]
    for row_dict in df.to_dict(orient="records"):

        p, r, f1 = bert_score.score(
            [row_dict[columns[0]]],
            [row_dict[columns[1]]],
            model_type=model,
            verbose=False
        )

        row_dict['bert_score'] = f1[0].item()
        dicts.append(row_dict)
        
    return pd.DataFrame(dicts)
```

```python
if run:
    correctness_bert_score_df = calculate_bert_score(qanda_responses_df, [settings['answer_column'], 'response'])
    correctness_bert_score_df.to_csv('output/correctness_bert_score.csv', index=False)
else:
    correctness_bert_score_df = pd.read_csv('output/correctness_bert_score.csv')

display(correctness_bert_score_df)
```

```python
correctness_by_model_df = correctness_bert_score_df.groupby('model_name')['bert_score'].mean()
display(correctness_by_model_df)
correctness_by_model_df.plot(
    kind='bar',
    title='correctness',
    xlabel='model',
    ylabel='bert score',
    ylim=[0.7, 0.8]
)
```

## Summarisation

```python
summaries_df = pd.read_parquet('data/validation-00000-of-00001.parquet')
summaries_df
```

```python
def summarise(df, models, settings):
    
    bedrock_runtime_client = boto3.client(service_name="bedrock-runtime", region_name='eu-west-2')
    
    dicts = []
    for row_dict in df.sample(n=settings['n_prompts']).to_dict(orient="records"):
        for model_name, model_id in models.items():
            print(f'prompting model {model_name}')
            row_dict['model_name'] = model_name
            row_dict['model_id'] = model_id
            messages = [{"role": "user", "content": [{"text": f"summarise: {row_dict[settings['document_column']]}"}]}]
            
            response = bedrock_runtime_client.converse(
                modelId=model_id,
                messages=messages,
                inferenceConfig={
                    "temperature": settings['temperature'],
                    "topP": settings['topP']
                }
            )
            
            row_dict['response'] = response['output']['message']['content'][-1]['text']   
            dicts.append(row_dict | response['usage'] | settings )

    return pd.DataFrame(dicts)
```

```python
if run:
    summarised_df = summarise(summaries_df, models, settings)
    summarised_df.to_csv('output/summarised.csv', index=False)
else:
    summarised_df = pd.read_csv('output/summarised.csv')

summarised_df
```

## Validate Summary with DeepEval

```python
# Without Human Summary

criteria = (
    "Summarization Quality: Evaluate if the summary is a concise, "
    "faithful, and coherent representation of the source document."
)
params = [LLMTestCaseParams.INPUT, LLMTestCaseParams.ACTUAL_OUTPUT]
expected_output=None
```

```python
def validate_summary_deepeval(df):

    criteria = (
        "Determine if the 'actual output' is factually correct and aligns "
        "with the key points presented in the 'expected output'."
    )
    params = [LLMTestCaseParams.ACTUAL_OUTPUT, LLMTestCaseParams.EXPECTED_OUTPUT]


    dicts = []
    for row_dict in df.to_dict(orient="records"):

        summarisation_metric = GEval(
            name="Summarization Quality",
            criteria=criteria,
            evaluation_params=params,
            threshold=0.7,
            async_mode=False
        )

        test_case = LLMTestCase(
            input=row_dict['document'],
            actual_output=row_dict['response'],
            expected_output=row_dict['summary']
        )

        row_dict['deepeval_score'] = summarisation_metric.measure(test_case)
        #row_dict['deepeval_reasoning'] = summarisation_metric.measure(test_case)
        #row_dict['deepeval_is_succesful'] = summarisation_metric.measure(test_case)

        dicts.append(row_dict)
        
    return pd.DataFrame(dicts)
```

```python
if run:
    summaries_deepeval_df = validate_summary_deepeval(summarised_df)
    summaries_deepeval_df.to_csv('output/summaries_deepeval.csv', index=False)
else:
    summaries_deepeval_df = pd.read_csv('output/summaries_deepeval.csv')

display(summaries_deepeval_df)
```

```python
summaries_deepeval_by_model_df = summaries_deepeval_df.groupby('model_name')['deepeval_score'].mean()
display(summaries_deepeval_by_model_df)
summaries_deepeval_by_model_df.plot(
    kind='bar',
    title='summarisation quality',
    xlabel='model',
    ylabel='deepeval score',
    ylim=[0.35, 0.65]
)
```

## Tuning

```python
def tune_parameter(models, parameter, values):
    dicts=[]
    for value in values:
        print(f'evaluating {parameter} = {value}:)
        dict = {}
        dict[parameter] = value
        settings[parameter] = value
        qanda_responses_temperature_df = get_qanda_response(qanda_df, models, settings)
        correctness_temperature_df = calculate_correctness(qanda_responses_temperature_df, settings)
        dict['deepeval_score'] = correctness_temperature_df['deepeval_score'].mean()
        dicts.append(dict)
        
    df = pd.DataFrame(dicts)  
    return df
```

## Tune Temperature

```python
models = {}
models['Claude 3.7 Sonnet'] = "anthropic.claude-3-7-sonnet-20250219-v1:0"
values = [0.1, 0.3, 0.5, 0.7, 0.9]
settings['topP'] = 0.3

if run:
    correctness_temperature_df = tune_parameter(models, 'temperature', values)
    correctness_temperature_df.to_csv('output/correctness_temperature.csv', index=False)
else:
    correctness_temperature_df = pd.read_csv('output/correctness_temperature.csv')
    
correctness_temperature_df = correctness_temperature_df.set_index('temperature')
correctness_temperature_df.plot(
    kind='line',
    title='correctness vs temperature',
    ylabel='deepeval score',
    legend=False
)
```

## Tune TopP

```python
values = [0.1, 0.3, 0.5, 0.7, 0.9]
settings['temperature'] = 0.5

if run:
    correctness_topP_df = tune_parameter(models, 'topP', values)
    correctness_topP_df.to_csv('output/correctness_topP.csv', index=False)
else:
    correctness_topP_df = pd.read_csv('output/correctness_topP.csv')
    
correctness_topP_df = correctness_topP_df.set_index('topP')
correctness_topP_df.plot(
    kind='line',
    title='correctness vs topP',
    ylabel='deepeval score',
    legend=False
)
```
